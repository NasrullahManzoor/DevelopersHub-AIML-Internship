# Task 1: Exploring and Visualizing a Simple Dataset

```python
# Importing necessary libraries for data analysis and visualization
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Setting the visual style for seaborn plots
sns.set_theme(style="whitegrid")
```

# Loading the Dataset

```python
# Loading the Iris dataset
df = sns.load_dataset('iris')

# Confirming data load
print("Dataset loaded successfully!")
```

# Data Inspection

```python
# Displaying the shape of the dataset (Rows, Columns)
print(f"Shape of the dataset: {df.shape}")

# Displaying the column names
print(f"\nColumns available:\n{df.columns.tolist()}")

# Displaying the first 5 rows to understand the data format
print("\nFirst 5 rows of the dataset:")
print(df.head())
```

# Statistical Summary

```python
# Checking for null values and data types
print("--- Data Info ---")
print(df.info())

# Generating a statistical summary
print("\n--- Statistical Description ---")
print(df.describe())
```

# Data Visualization

## 1 :  Scatter Plot: To see the relationship between features.

```python
# A) Scatter Plot: Relationship between Sepal Length and Petal Length
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species', s=100)
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.legend(title='Species')
plt.show()
```

## 2 : Histograms: To see the distribution of data.

```python
# B) Histograms: Checking the distribution of each feature
# This helps us see if the data is normally distributed or skewed
df.hist(figsize=(12, 8), color='teal', edgecolor='black', alpha=0.7)
plt.suptitle("Histograms of Feature Distributions", fontsize=16)
plt.show()
```

## 3 : Box Plots: To identify outliers.

```python
# C) Box Plots: detecting Outliers
# Points outside the 'whiskers' of the box are considered outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, palette="Set2")
plt.title("Box Plots for Outlier Detection")
plt.show()
```
